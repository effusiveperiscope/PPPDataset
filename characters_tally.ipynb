```python
from ppp import PPPDataset
data = PPPDataset.collect([])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

char_dict = {}
for char,files in data.file_dict.items():
    char_dict[char] = len(files)

sorted_char_list = sorted(char_dict.items(), key=lambda x: x[1], reverse=True)
magnitude_lists = {1000:[], 100:[], 10:[], 1:[]}

for tup in sorted_char_list:
    k,v = tup
    magnitude = 10 ** (len(str(v)) - 1)
    magnitude_lists[magnitude].append(tup)

for magnitude, items in magnitude_lists.items():
    k,v = zip(*magnitude_lists[magnitude])
    colors = plt.cm.magma(np.linspace(0, 1, len(k)))

    df = pd.DataFrame(items, columns=['character','line_count'])
    df = df[::-1]
    #plt.figure(figsize=(8,3))
    #plt.axis('off')
    #plt.table(cellText=df.values, colLabels=df.columns, loc='center', cellLoc='center')
    #plt.savefig(f'PPPData_{int(magnitude)}s_table.png', bbox_inches='tight', pad_inches=1)

    plt.figure(figsize=(7,len(k)))
    bars = plt.barh(df['character'],df['line_count'],color=colors, height=0.8)
    for bar, value in zip(bars, df['line_count']):
        plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value}', ha='left', va='center')

    plt.ylabel('Characters')
    plt.xlabel('Line count')
    plt.savefig(f'PPPData_{int(magnitude)}s.png', bbox_inches='tight', pad_inches=1)
```

```python
from ppp import PPPDataset
data = PPPDataset.collect([])
char_dict = data.stats()

def fmt_seconds(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    return f"{int(hours):02d}:{int(minutes):02d}:{int(seconds):02d}"
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sorted_char_list = sorted(char_dict.items(), key=lambda x: x[1], reverse=True)
magnitude_lists = {10000: [], 1000:[], 100:[], 10:[], 1:[]}

for tup in sorted_char_list:
    k,v = tup
    print(str(v))
    magnitude = 10 ** (len(str(int(v))) - 1)
    magnitude_lists[magnitude].append(tup)

for magnitude, items in magnitude_lists.items():
    k,v = zip(*magnitude_lists[magnitude])
    colors = plt.cm.magma(np.linspace(0, 1, len(k)))

    df = pd.DataFrame(items, columns=['character','duration'])
    df = df[::-1]

    plt.figure(figsize=(7,len(k)))
    bars = plt.barh(df['character'],df['duration'],color=colors, height=0.8)
    for bar, value in zip(bars, df['duration']):
        plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, fmt_seconds(value), ha='left', va='center')

    plt.ylabel('Characters')
    plt.xlabel('Duration')
    plt.savefig(f'PPPData_dur_{int(magnitude)}s.png', bbox_inches='tight', pad_inches=1)
```
